```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv('amz_uk_price_prediction_dataset.csv')

df


frequency_table = df['category'].value_counts()
frequency_table

```

```python

frequency_table = df['category'].value_counts()
top_10_categories = frequency_table.head(10)


plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_categories.index, y=top_10_categories.values, palette="Set3")
plt.xticks(rotation=90)
plt.title('Distribution of Products Across Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Number of Listings')
plt.show()


top_5_categories = frequency_table.head(5)
plt.figure(figsize=(12, 8))
plt.pie(top_5_categories, labels=top_5_categories.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.title('Top 5 Most Listed Product Categories')
plt.ylabel('')
plt.show()


mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")


variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)
print(f"Variance of Price: {variance_price}")
print(f"Standard Deviation of Price: {std_dev_price}")
print(f"Range of Price: {range_price}")
print(f"Interquartile Range of Price: {iqr_price}")

plt.figure(figsize=(12, 8))
sns.histplot(np.log1p(df['price']), kde=True, bins=30, color="salmon")
plt.title('Distribution of Product Prices (Log-Transformed)')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(12, 8))
sns.boxplot(x=np.log1p(df['price']), color="lightblue")
plt.title('Box Plot of Product Prices (Log-Transformed)')
plt.xlabel('Log(Price)')
plt.show()


mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]
print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")


variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)
print(f"Variance of Rating: {variance_rating}")
print(f"Standard Deviation of Rating: {std_dev_rating}")
print(f"Interquartile Range of Rating: {iqr_rating}")


skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()
print(f"Skewness of Rating: {skewness_rating}")
print(f"Kurtosis of Rating: {kurtosis_rating}")

plt.figure(figsize=(12, 8))
sns.histplot(df['stars'], kde=True, bins=30, color="salmon")
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()


df['price'] = df['price'].dropna()
df['stars'] = df['stars'].dropna()
```
